```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
```

```python
import seaborn as sb
sb.set_style("whitegrid")
```

### Eyeballing dataset distributions with histograms

```python
address = 'C:/Users/yojit/Documents/Python/5_Python_for_Data_Science_E/Data/mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

cars.index = cars.car_names

mpg = cars.mpg
```

```python
mpg.plot(kind = "hist")
```

```python
plt.hist(mpg)
plt.show()
```

```python
sb.displot(mpg)
```

###  Seeing scatterplots in action

```python
cars.plot(kind = "scatter", x = "hp", y = "mpg", c = ["red"], s = 50, alpha = 0.5)
```

```python
sb.regplot(data = cars, x = "hp", y = "mpg", scatter = True)
```

```python
sb.regplot(data = cars, x = "hp", y = "mpg", scatter = False)
```

### Generating a scatter plot matrix

```python
sb.pairplot(cars)
```

```python
cars_subset = cars[["mpg", "disp", "hp", "wt"]]
sb.pairplot(cars_subset, palette = "hls")
plt.show()
```

### Building boxplots

```python
sb.color_palette("cubehelix", as_cmap=True)
```

```python
cars.boxplot(column = "mpg", by = "am")
cars.boxplot(column = "wt", by = "am")
```

```python
sb.boxplot(x="am", y = "mpg", data = cars, palette = "hls")
```
